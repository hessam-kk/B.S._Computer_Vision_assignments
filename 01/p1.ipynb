```python
# 2 - 2 - c

import cv2 
import math
import numpy as np
import matplotlib.pyplot as plt    

def imadjust(x,a,b,c,d,gamma=1):
    # Similar to imadjust in MATLAB.
    # Converts an image range from [a,b] to [c,d].
    # The Equation of a line can be used for this transformation:
    #   y=((d-c)/(b-a))*(x-a)+c
    # However, it is better to use a more generalized equation:
    #   y=((x-a)/(b-a))^gamma*(d-c)+c
    # If gamma is equal to 1, then the line equation is used.
    # When gamma is not equal to 1, then the transformation is not linear.

    y = (((x - a) / (b - a)) ** gamma) * (d - c) + c
    return y


# Reading
img7 = plt.imread('bridge2.gif')

print('Pixels read: {}'.format(img7.shape))
print('Type of data: {}'.format(img7.dtype))

fig = plt.figure(figsize=(16, 10), dpi=80)
# ax1, ax2 = fig.subplots(1, 3)
(ax1, ax2, ax3), (ax4, ax5, ax6) = fig.subplots(2, 3)


ax1.imshow(img7, cmap='gray')
ax1.set_title('Original Image')


stretched_image1 = imadjust(img7, img7.min(), img7.max(), 0, 255, 1)
stretched_image2 = imadjust(img7, img7.min(), img7.max(), 0, 255, 2)
stretched_image3 = imadjust(img7, img7.min(), img7.max(), 0, 255, 3)
stretched_image4 = imadjust(img7, img7.min(), img7.max(), 0, 255, 4)
stretched_image5 = imadjust(img7, img7.min(), img7.max(), 0, 255, 5)


ax2.imshow(stretched_image1, cmap='gray')
ax2.set_title('Stretched Image | Gamma 1')
ax3.imshow(stretched_image2, cmap='gray')
ax3.set_title('Stretched Image | Gamma 2')
ax4.imshow(stretched_image3, cmap='gray')
ax4.set_title('Stretched Image | Gamma 3')
ax5.imshow(stretched_image4, cmap='gray')
ax5.set_title('Stretched Image | Gamma 4')
ax6.imshow(stretched_image5, cmap='gray')
ax6.set_title('Stretched Image | Gamma 5')


print('Original Image: <min: {}>, <max: {}>, <mean: {}>, <std: {}>\n'.format(
            img7.min(), img7.max(), img7.mean(), img7.std()
))

print('Stretched Image 1 Image: <min: {}>, <max: {}>, <mean: {}>, <std: {}>\n'.format(
            stretched_image1.min(), stretched_image1.max(), stretched_image1.mean(), stretched_image1.std()
))

print('Stretched Image 2 Image: <min: {}>, <max: {}>, <mean: {}>, <std: {}>\n'.format(
            stretched_image2.min(), stretched_image2.max(), stretched_image2.mean(), stretched_image2.std()
))

print('Stretched Image 3 Image: <min: {}>, <max: {}>, <mean: {}>, <std: {}>\n'.format(
            stretched_image3.min(), stretched_image3.max(), stretched_image3.mean(), stretched_image3.std()
))

print('Stretched Image 4 Image: <min: {}>, <max: {}>, <mean: {}>, <std: {}>\n'.format(
            stretched_image4.min(), stretched_image4.max(), stretched_image4.mean(), stretched_image4.std()
))

print('Stretched Image 5 Image: <min: {}>, <max: {}>, <mean: {}>, <std: {}>\n'.format(
            stretched_image5.min(), stretched_image5.max(), stretched_image5.mean(), stretched_image5.std()
))

```
